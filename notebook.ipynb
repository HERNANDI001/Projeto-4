### **Tratamento de Dados**

#### **Juntando as duas tabelas**


```python
tabela = pd.concat([etanol11,etanol12], ignore_index=True)
tabela
```

#### **Retirando:**


- CEP;
- CNPJ;
- Valor de COmpra;
- Unidade de Medida;
- Complemento;
- Numero Rua;
- Nome da Rua;

```python
tabela.drop(['Cep','CNPJ da Revenda','Valor de Compra','Unidade de Medida','Complemento','Numero Rua','Nome da Rua'], inplace=True, axis='columns')
tabela
```

#### **Reordenando rotulos**

```python
tabela = tabela.reindex(columns=['Regiao - Sigla','Estado - Sigla','Municipio','Bairro','Bandeira','Revenda','Valor de Venda','Produto','Data da Coleta'])
tabela
```

### **Separando em tabelas por produto**

#### Etanol

```python
etanol = tabela.loc[tabela['Produto'] == 'ETANOL']
etanol
```

#### Gasolina

```python
gasolina = tabela.loc[tabela['Produto'] == 'GASOLINA']
gasolina
```

#### Gasolina Aditivada

```python
gasolina_a = tabela.loc[tabela['Produto'] == 'GASOLINA ADITIVADA']
gasolina_a
```
