### **Tratamento de Dados**

#### **Juntando as duas tabelas**

#### **Retirando:**


- CEP;
- CNPJ;
- Valor de COmpra;
- Unidade de Medida;
- Complemento;
- Numero Rua;
- Nome da Rua;

```python
tabela.drop(['Cep','CNPJ da Revenda','Valor de Compra','Unidade de Medida','Complemento','Numero Rua','Nome da Rua'], inplace=True, axis='columns')
tabela
```

#### **Reordenando rotulos**

```python
tabela = tabela.reindex(columns=['Regiao - Sigla','Estado - Sigla','Municipio','Bairro','Bandeira','Revenda','Valor de Venda','Produto','Data da Coleta'])
tabela
```

#### **Mudando o Formato de Data e Preço**

```python
tabela['Data da Coleta'] = pd.to_datetime(tabela['Data da Coleta'], format='%d/%m/%Y')
tabela['Valor de Venda'] = tabela['Valor de Venda'].str.replace(',', '.').astype(float)
tabela
```
